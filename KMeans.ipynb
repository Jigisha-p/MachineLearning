```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from matplotlib import pyplot
```

```python
# create a dataset using make_blobs()
dt,y=make_blobs(n_samples=240, centers=4,cluster_std=0.60)
```

```python
# Data visualization
X1= dt[:,0]
X2= dt[:,1]
pyplot.scatter(X1,X2)
```

```python
c=KMeans(n_clusters=4)
c.fit(dt)
```

```python
pred=c.predict(dt)
pyplot.scatter(X1,X2,c=pred)
centers=c.cluster_centers_
pyplot.scatter(centers[:,0],centers[:,1],marker='x',c='white',s=80)
```

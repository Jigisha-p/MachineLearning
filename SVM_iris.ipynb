```python
import numpy as np 
from sklearn import svm,datasets  
from matplotlib import pyplot 

```

```python
dt=datasets.load_iris()
X=dt.data[:,:2]
y=dt.target
```

Now we need the data to plot the SVM boundaries  or input data spaces(different classes). To do so  we need to create a grid. To plot the grid, we need minimum and maximum values of the  input and output datasets. Then we reshape them  using ravel() and pass them to the c_() function  which stacks arrays along their last  axis after being upgraded to at least 2-D and  stored in the X_plot as testing input 
 

```python
x_min, x_max=X[:,0].min() -1, X[:,0].min() +1
y_min, y_max=X[:,1].min() -1, X[:,1].min() +1
h=(x_max /x_min)/100
xx,yy=np.meshgrid(np.arange(x_min,x_max,h),
                  np.arange(y_min,y_max,h))
X_plot=np.c_[xx.ravel(),yy.ravel()]
```

```python
SvcModel=svm.SVC(kernel='linear',C=1.0).fit(X,y)  # C is a regularization parameter
```

Now we can predict the X_plot values and store  it in Z. We will reshape it using the reshape()  function to the shape of the xx meshgrid  First of all we will plot the figure(base).


```python
Z=SvcModel.predict(X_plot)
Z=Z.reshape(xx.shape)
pyplot.figure(figsize=(15,5))
pyplot.subplot(121)
pyplot.contourf(xx,yy,Z,alpha=0.3)
pyplot.scatter(X[:,0],X[:,1],c=y)
pyplot.xlim(xx.min(),xx.max())
```

Similarly we can create a SVC using the Radial  Basis Function kernel. 



```python
RbfSvc=svm.SVC(kernel='rbf',C=1.0).fit(X,y)
Z=RbfSvc.predict(X_plot)
Z=Z.reshape(xx.shape)
pyplot.figure(figsize=(15,5))
pyplot.subplot(121)
pyplot.contourf(xx,yy,Z,alpha=0.3)
pyplot.scatter(X[:,0],X[:,1],c=y)
pyplot.xlim(xx.min(),xx.max())
```

You can observe both of the plots using the  linear and rbf and notice and notice a clear  difference in lines and curves
